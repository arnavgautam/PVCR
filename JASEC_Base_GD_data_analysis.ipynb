```julia
using DataFrames
using CSV
include("functions/plotting_base_gd.jl")
```

```julia
# Read in all of the utility bills for the various utilities and tariff types
base_GD = CSV.read("data/Base_GD.csv")

# Narrow down to {Residential | Commerical | etc} JASEC customers

all_jasec = filter(row -> (!ismissing(row.EMPRESA) && row.EMPRESA == "JASEC"), base_GD)
all_jasec_tariff_split = Array{DataFrame}(undef, 12)

# Loop through all tariffs
for i=1:12
    next_jasec_value = filter(row -> (!ismissing(row.CODIGO_TARIFA) && row.CODIGO_TARIFA == string(i)), all_jasec)
    all_jasec_tariff_split[i] = next_jasec_value
end
```

```julia
plot_all_tariffs_per_company(all_jasec_tariff_split, "JASEC")
```
