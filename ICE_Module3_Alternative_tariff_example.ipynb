## Change in residential tariff with new rules. Case with No PV
### Consumer 550 kWh/month

```julia
PVRes.capacity = 0.0
ice_residential.econsumption = 550; get_pmax(ice_residential);
ebalance = annual_energy_balance(ice_residential, PVRes, print_output=false);
```

```julia
bills_residential1 = Dict()
for m in 1:12
    bills_residential1[m] = monthly_bill(ebalance[m], ice_residential, print_output=false)
end
```

```julia
bills_residential2 = Dict()
ice_residential.tariff = A_R_ICE
for m in 1:12
    bills_residential2[m] = monthly_bill(ebalance[m], ice_residential; print_output=false, tax_cutoff = 0.0)
end
```

```julia
plot_bill(bills_residential1, bills_residential2, ["grid_cost", "demand_charge", "VAT", "street_light", "firefighters"]);
```
