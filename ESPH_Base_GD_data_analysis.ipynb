```julia
using DataFrames
using CSV
include("functions/plotting_base_gd.jl")
```

```julia
# Read in all of the utility bills for the various utilities and tariff types
base_GD = CSV.read("data/Base_GD.csv")

# Narrow down to {Residential | Commerical | etc} ESPH customers

all_esph = filter(row -> (!ismissing(row.EMPRESA) && row.EMPRESA == "ESPH"), base_GD)
all_esph_tariff_split = Array{DataFrame}(undef, 12)

# Loop through all tariffs
for i=1:12
    next_esph_value = filter(row -> (!ismissing(row.CODIGO_TARIFA) && row.CODIGO_TARIFA == string(i)), all_esph)
    all_esph_tariff_split[i] = next_esph_value
end
```

```julia
plot_all_tariffs_per_company(all_esph_tariff_split, "ESPH")
```
