```julia
using DataFrames
using CSV
include("functions/plotting_base_gd.jl")
```

```julia
# Read in all of the utility bills for the various utilities and tariff types
base_GD = CSV.read("data/Base_GD.csv")

# Narrow down to {Residential | Commerical | etc} {CNFL | ICE | ESPH | JASEC} customers

all_cnfl = filter(row -> (!ismissing(row.EMPRESA) && row.EMPRESA == "CNFL"), base_GD)
all_cnfl_tariff_split = Array{DataFrame}(undef, 12)

all_ice = filter(row -> (!ismissing(row.EMPRESA) && row.EMPRESA == "ICE"), base_GD)
all_ice_tariff_split = Array{DataFrame}(undef, 12)

all_esph = filter(row -> (!ismissing(row.EMPRESA) && row.EMPRESA == "ESPH"), base_GD)
all_esph_tariff_split = Array{DataFrame}(undef, 12)

all_jasec = filter(row -> (!ismissing(row.EMPRESA) && row.EMPRESA == "JASEC"), base_GD)
all_jasec_tariff_split = Array{DataFrame}(undef, 12)

for i=1:12
    next_cnfl_value = filter(row -> (!ismissing(row.CODIGO_TARIFA) && row.CODIGO_TARIFA == string(i)), all_cnfl)
    all_cnfl_tariff_split[i] = next_cnfl_value
    
    next_ice_value = filter(row -> (!ismissing(row.CODIGO_TARIFA) && row.CODIGO_TARIFA == string(i)), all_ice)
    all_ice_tariff_split[i] = next_ice_value
    
    next_esph_value = filter(row -> (!ismissing(row.CODIGO_TARIFA) && row.CODIGO_TARIFA == string(i)), all_esph)
    all_esph_tariff_split[i] = next_esph_value
    
    next_jasec_value = filter(row -> (!ismissing(row.CODIGO_TARIFA) && row.CODIGO_TARIFA == string(i)), all_jasec)
    all_jasec_tariff_split[i] = next_jasec_value
end
```

```julia
plot_all_tariffs_per_company(all_cnfl_tariff_split, "CNFL")
```

```julia
plot_all_tariffs_per_company(all_ice_tariff_split, "ICE")
```

```julia
plot_all_tariffs_per_company(all_esph_tariff_split, "ESPH")
```

```julia
plot_all_tariffs_per_company(all_jasec_tariff_split, "JASEC")
```
